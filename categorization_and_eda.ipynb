```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes_data.csv')
df.head(5)
```

```python
df.describe()
```

```python
df = df.drop(columns=['PatientID', 'DoctorInCharge'])
df.shape
```

```python
df.dtypes
```

```python
categorical_identifier = [0, 1, 2, 3]
categorical_vars = []
continuous_vars = []

for col in df.columns:
    if df[col].max() in categorical_identifier:
        categorical_vars.append(col)
    else:
        continuous_vars.append(col)

print("Categorical Variables:")
print(categorical_vars)

print("\nContinuous Variables:")
print(continuous_vars)
```

```python
binary_categoricals = [col for col in categorical_vars if df[col].nunique() == 2]
multi_categoricals = [col for col in categorical_vars if 2 < df[col].nunique() <= 10]

print("Binary Categorical Variables:")
print(binary_categoricals)
print("\nMulti Categorical Variables:")
print(multi_categoricals)
```

```python
def group_and_plot_continuous_variables(df, plot=True):
    grouped_vars = {
        'demographic_behavioral': [
            'Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity',
            'DietQuality', 'SleepQuality', 'MedicalCheckupsFrequency',
            'HealthLiteracy', 'QualityOfLifeScore'
        ],
        'clinical_lab': [
            'SystolicBP', 'DiastolicBP', 'FastingBloodSugar', 'HbA1c',
            'SerumCreatinine', 'BUNLevels', 'CholesterolTotal',
            'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides'
        ],
        'symptom_subjective': [
            'FatigueLevels'
        ],
        'target': ['MedicationAdherence']
    }

    if plot:
        for category, variables in grouped_vars.items():
            print(f"\n📊 Plotting {category.replace('_', ' ').title()} Variables")
            n = len(variables)
            rows = (n + 2) // 3  # ensures up to 3 columns per row
            fig, axes = plt.subplots(rows, 3, figsize=(16, 4 * rows))
            axes = axes.flatten()
            
            for i, var in enumerate(variables):
                if var in df.columns:
                    sns.histplot(df[var].dropna(), bins=20, kde=True, ax=axes[i])
                    axes[i].set_title(var)
                else:
                    axes[i].set_visible(False)

            # Hide any unused subplots
            for j in range(i + 1, len(axes)):
                axes[j].set_visible(False)

            plt.tight_layout()
            plt.show()

    return grouped_vars
```

```python
group_and_plot_continuous_variables(df, plot=True)
```

```python
def group_and_plot_categorical_variables(df, plot=True):
    grouped_categoricals = {
        'demographics': [
            'Gender', 'Ethnicity', 'SocioeconomicStatus', 'EducationLevel'
        ],
        'lifestyle_history': [
            'Smoking', 'FamilyHistoryDiabetes', 'GestationalDiabetes',
            'PolycysticOvarySyndrome', 'PreviousPreDiabetes', 'Hypertension'
        ],
        'diagnosis_and_medications': [
            'AntihypertensiveMedications', 'Statins', 'AntidiabeticMedications', 'Diagnosis'
        ],
        'symptoms_quality': [
            'FrequentUrination', 'ExcessiveThirst', 'UnexplainedWeightLoss',
            'BlurredVision', 'SlowHealingSores', 'TinglingHandsFeet'
        ],
        'environmental_exposures': [
            'HeavyMetalsExposure', 'OccupationalExposureChemicals', 'WaterQuality'
        ]
    }

    if plot:
        for category, variables in grouped_categoricals.items():
            print(f"\n📊 Plotting {category.replace('_', ' ').title()} Variables")
            n = len(variables)
            rows = (n + 2) // 3  # up to 3 columns per row
            fig, axes = plt.subplots(rows, 3, figsize=(16, 4 * rows))
            axes = axes.flatten()

            for i, var in enumerate(variables):
                if var in df.columns:
                    sns.countplot(x=var, data=df, ax=axes[i])
                    axes[i].set_title(var)
                    axes[i].tick_params(axis='x', rotation=45)
                else:
                    axes[i].set_visible(False)

            for j in range(i + 1, len(axes)):
                axes[j].set_visible(False)

            plt.tight_layout()
            plt.show()

    return grouped_categoricals
```

```python
group_and_plot_categorical_variables(df, plot=True)
```
